```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pyupbit
from sklearn import preprocessing
```

## 상장일-현재

```python
ticker = 'KRW-NU'
interval = 'day'
to = '2021-11-16'
nucypher = pd.DataFrame(pyupbit.get_ohlcv(ticker=ticker,interval=interval,to=to))
pd.options.display.float_format = '{:.2f}'.format
nucypher.reset_index(inplace=True)
nucypher = nucypher.rename(columns={'index':'date'})
nucypher
```

```python
#nucyper.to_csv("nucyper_211015_211115.csv")
```

```python
plt.figure(figsize=(30, 15))
plt.title("Nucyper_210115_211115", fontsize=40)
plt.plot(nucypher.date, nucypher['open'], "#81C349")
plt.plot(nucypher.date, nucypher['close'], "#FFA500")
plt.plot(nucypher.date, nucypher['high'], "#B7142D")
plt.plot(nucypher.date, nucypher['low'], "#1259BF")
plt.xlabel('date', fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(['open', 'close', 'high', 'low'], prop={'size': 25})
plt.show()
```

```python
nucypher = nucypher.astype({'open': int, 'high': int, 'low': int, 'close':int})
nucypher.drop(nucypher.columns[[5,6]], axis=1, inplace=True)
nucypher
```

## 한달거래량과 종가

```python
nu_sample = pd.read_csv("data/최종/nucyper_211015_211115.csv", index_col=0)
n_sample = nu_sample.drop(nu_sample.columns[0], axis=1)
nu_sample= nu_sample.set_index("date")
nu_sample = nu_sample.drop(["open","high","low", "value"], axis=1)
nu_sample.index = nu_sample.index.str.replace(" 09:00:00", "")


pd.options.display.float_format = '{:.2f}'.format
nu_sample
```

```python
#nu_sample.to_csv("data/final/nucypher_close_volume.csv")
```

```python
from matplotlib import dates

plt.rcParams['figure.figsize'] = (15, 8)
plt.rcParams['font.size'] = (15)

x = nu_sample.index
y1 = nu_sample.close
y2 = nu_sample.volume

fig, ax1 = plt.subplots()
ax1.xaxis.set_major_locator(dates.DayLocator(interval=4))
plt.xticks(rotation=45)
ax1.plot(x, y1, color="#db58a9")
ax1.set_ylabel('Close')
ax1.set_xlabel('Date')
ax1.tick_params(axis='x', labelsize=10)
ax1.tick_params(axis='y', labelsize=10)

ax2 = ax1.twinx()
ax2.plot(x, y2, label="volume", linewidth=0)
ax2.set_ylabel('Volume')
ax2.tick_params(axis='y', labelsize=10)
plt.fill_between(x, y2, 0, alpha=0.2, facecolor='grey', antialiased=True)


ax1.set_zorder(ax1.get_zorder() + 10)
ax1.patch.set_visible(False)
plt.title("nucyper_1month_close_volume")
plt.show()
```
